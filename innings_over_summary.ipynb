```python
import matplotlib.pyplot as plt

def plot_cricket_bars(overs, runs, dot_positions=None, title="Cricket Innings"):
    """
    Plots bar chart with dots (e.g., for wickets) above each bar.
    
    Parameters:
    - overs: List of over numbers [1, 2, ..., 20]
    - runs: List of runs scored in each over
    - dot_positions: Dictionary {over_number: num_dots}
    """
    fig, ax = plt.subplots(figsize=(12, 6))

    bars = ax.bar(overs, runs, color='red', edgecolor='black', alpha=0.8)

    # Plot dots above bars
    if dot_positions:
        for i, over in enumerate(overs):
            num_dots = dot_positions.get(over, 0)
            for j in range(num_dots):
                ax.plot(
                    over,
                    runs[i] + 1 + j * 1.2,
                    'ko',  # black circle marker
                    markersize=8
                )

    ax.set_title(title, fontsize=14)
    ax.set_xlabel("Overs", fontsize=12)
    ax.set_ylabel("Runs", fontsize=12)
    ax.set_xticks(overs)
    ax.grid(True, linestyle='--', alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
overs = list(range(1, 21))
runs = [6, 11, 9, 6, 5, 6, 10, 9, 12, 6, 13, 6, 5, 11, 15, 17, 16, 7, 14, 1]
wickets = {
    2: 1,
    5: 1,
    8: 2,
    13: 1,
    16: 2,
    19: 1,
    20: 1
}

plot_cricket_bars(overs, runs, dot_positions=wickets, title="Warwickshire Innings")
```
